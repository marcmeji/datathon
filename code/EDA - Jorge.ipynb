```python
plt.figure(figsize=(17, 10))
sns.heatmap(df.corr(), annot=True, annot_kws={"size": 7}, linewidths=1)
plt.show()
```

### Distribución y diagrama: sugar_index

```python
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.40, 1)})
 
# Plot
sns.boxplot(df.sugar_index[df.sugar_index.notnull()], ax=ax_box)
sns.distplot(df.sugar_index[df.sugar_index.notnull()], ax=ax_hist, bins=10)
plt.show()
```

Quitar outliers de sugar_index

```python
Q1 = df['sugar_index'].quantile(0.25)
Q3 = df['sugar_index'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5* IQR
upper = Q3 + 1.5* IQR

df = df.loc[(df['sugar_index'] > lower) & (df['sugar_index'] < upper)]

f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.40, 1)})
 
# Plot
sns.boxplot(df.sugar_index[df.sugar_index.notnull()], ax=ax_box)
sns.distplot(df.sugar_index[df.sugar_index.notnull()], ax=ax_hist, bins=10)
plt.show()
```
