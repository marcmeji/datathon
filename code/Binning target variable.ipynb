```python
import pandas as pd
```

```python
df = pd.read_csv('../data/cookies_FS.csv')
```

```python
df.head()
```

```python
low = [3,4,5]
medium = [6,7,8]
high = [9,10,11]

def bins (x):
    if x in low:
        return 0
    elif x in medium:
        return 1
    elif x in high:
        return 2

df['quality_binned'] = df['quality'].apply(bins)
```

```python
df = df.drop(columns='quality')
```
